```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = data.drop(['Cabin'],axis=1)
```

```python
data = data.drop(['PassengerId','Name','Ticket'],axis=1)
```

```python
data.columns
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
mapping = {'male':1,'female':0}
data['Sex'] = data['Sex'].map(mapping)
```

```python
mapping = {'C': 0,'Q': 1,'S': 2}
data['Embarked'] = data['Embarked'].map(mapping)
```

```python
sns.countplot(x='Survived',data=data)
plt.title('Survival Count')
plt.xticks([0,1],['Did not survive','survived'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survived by Passenger Class')
plt.legend(title='Suevived', labels=['No','Yes'])
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived',data=data)
plt.title('Survival by Sex')
plt.legend(title='Survived',labels=['No','Yes'])
plt.show()
```

```python
sns.histplot(data['Age'], kde=True, bins=40)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(x='Survived', y='Fare', data=data)
plt.title('Fare Paid vs Survival')
plt.xticks([0,1 ],['Did not survive','Survived'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:",accuracy_score(y_test,y_pred))
print("\n Classification Report :\n ", classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',xticklabels=["Not Survived","Survived"],yticklabels=["Not Survived","Survived"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
